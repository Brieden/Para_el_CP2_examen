```python
n = 1000
stochastic_variabel = np.floor(np.random.rand(n)*6+1)
# Prozentuale Verteilung
if 0:
    for i in range(1,7):
        print(i, sum(stochastic_variabel==i)/n*100)
# Absolute Verteilung der Augenzahl
if 1:
    plt.hist(stochastic_variabel,bins=6, range=(min(stochastic_variabel),max(stochastic_variabel)))
    plt.show()
# Errechnen der Paare
pair_matrix = np.zeros([6,6])
for ni, i in enumerate(stochastic_variabel):
    for nj, j in enumerate(stochastic_variabel[ni+1:]):
        pair_matrix[int(i)-1,int(j)-1] += 1
pair_einzeln_matrix = np.triu(pair_matrix + pair_matrix.T - pair_matrix * np.eye(6))
print(pair_matrix, np.sum(pair_matrix))
print(pair_einzeln_matrix, np.sum(pair_einzeln_matrix))
plt.plot(pair_matrix.flatten(), ".")
plt.ylim(0, pair_matrix.max())
plt.show()
# 
pair_einzeln_liste = []
for i in np.arange(1,7):
    for j in np.arange(1,7):
        for k in range(int(pair_einzeln_matrix[int(i)-1,int(j)-1])):
            pair_einzeln_liste.append(i + 0.1 * j)
if 1:
    plt.hist(pair_einzeln_liste, bins = 300)
    plt.show()   
print([pair_matrix[i,i] for i in np.arange(0,6)])
```

```python
def wurf():
    wÃ¼rfel1 = randint(1, 6)
    wÃ¼rfel2 = randint(1, 6)
    if wÃ¼rfel1 >= wÃ¼rfel2:
        result = wÃ¼rfel1 * 10 + wÃ¼rfel2
    else:
        result = wÃ¼rfel2 * 10 + wÃ¼rfel1
    return result

def vergleich(result1, result2):
    rank = [31, 32, 41, 42, 43, 51, 52, 53, 54, 61, 62, 63, 64, 65, 11, 22, 33, 44, 55, 66, 21]
    return bool(rank.index(result1) >= rank.index(result2))
    #return bool(rank.index(result1) > rank.index(result2)) - bool(rank.index(result1) < rank.index(result2))

durations = []
t = 0
a = wurf()
b = wurf()
for i in range(1000000):
    if vergleich(a,b):
        t += 1
    else:
        durations.append(t)
        t =0
    b = a
    a = wurf()
```

```python
plt.hist(durations, range=(0,max(durations)+1))
plt.yscale('log', nonposy='clip')
plt.show()
```
